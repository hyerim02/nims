```python
titanic_preprocessed = pd.read_csv('titanic.csv.gz')

titanic_preprocessed = titanic_preprocessed.set_index('Unnamed: 0')
titanic_preprocessed = titanic_preprocessed.drop(columns=['sibsp', 'deck', 'alone'])
titanic_preprocessed['embarked'] = titanic_preprocessed['embarked'].astype('category')
titanic_preprocessed['fare_grade'] = pd.cut(
    titanic_preprocessed['fare'],
    bins=[-np.inf, 10, 15, 30, np.inf],
    labels=['A', 'B', 'C', 'D'],
)
titanic_preprocessed['uppercase_who'] = titanic_preprocessed['who'].str.upper()

display(titanic_preprocessed)
titanic_preprocessed.info()
```

```python
titanic_preprocessed = (
    pd.read_csv('titanic.csv.gz')
    .set_index('Unnamed: 0')
    .drop(columns=['sibsp', 'deck', 'alone'])
    .assign(
        embarked=titanic_preprocessed['embarked'].astype('category'),
        fare_grade=pd.cut(
            titanic_preprocessed['fare'],
            bins=[-np.inf, 10, 15, 30, np.inf],
            labels=['A', 'B', 'C', 'D'],
        ),
        uppercase_who=titanic_preprocessed['who'].str.upper(),
    )
)

display(titanic_preprocessed)
titanic_preprocessed.info()
```

```python
titanic_preprocessed = pd.read_csv(
    'titanic.csv.gz',
    index_col=0,
    dtype={
        'embarked': 'category',
    },
)
titanic_preprocessed
```
